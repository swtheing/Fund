```python
# -*- coding:utf-8 -*-
import akshare as ak
fund_rating_all_df = ak.fund_rating_all()
print(fund_rating_all_df)
fund_em_open_fund_info_df = ak.fund_em_open_fund_info(fund="001487", indicator="累计净值走势")
print(fund_em_open_fund_info_df)
fund_portfolio_hold_em_df = ak.fund_portfolio_hold_em(code="001487", year="2021")
print(fund_portfolio_hold_em_df)
```

```python
hold_em_dic = {}
for fund_id in fund_rating_all_df["代码"]:
    try:
        print(len(hold_em_dic))
        fund_portfolio_hold_em_df = ak.fund_portfolio_hold_em(code=fund_id, year="2021")
        hold_em_dic[fund_id] = fund_portfolio_hold_em_df["股票名称"]
    except:
        continue
print(len(hold_em_dic))
```

```python
#print(hold_em_dic)
from collections import defaultdict
import numpy as np
save_dic = defaultdict(list)
for key, value in hold_em_dic.items():
    for val in value:
        save_dic[key].append(val)
np.save("hold_em_dic.npy", save_dic)
print(save_dic['560006'])
```

```python
stock = {}
count = 0
for key, value in hold_em_dic.items():
    for val in value:
        if val not in stock:
            stock[val] = count
            count += 1
print(stock)
```

```python
import numpy as np
features = np.zeros((len(hold_em_dic), len(stock)))
name_dic = {}
count = 0
for key, value in hold_em_dic.items():
    name_dic[count] = key
    for val in value:
        features[count][stock[val]] = 1
    count += 1
print(features)
```

```python
n_dic = {}
for i, key in enumerate(fund_rating_all_df["代码"]):
    n_dic[key] = fund_rating_all_df["简称"][i]
print(n_dic)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
from sklearn.cluster import KMeans
y_pred = KMeans(n_clusters=30, random_state=9).fit_predict(features)
cluster = defaultdict(list)
cluster_name = defaultdict(list)
for i, y in enumerate(y_pred):
    cluster_name[y].append(n_dic[name_dic[i]])
    cluster[y].append(name_dic[i])
print(cluster)
```

```python
fp = open("cluster_name.csv", "w")
for key, value in cluster_name.items():   
    fp.write(",".join(value) + "\n")
fp.close()
```
