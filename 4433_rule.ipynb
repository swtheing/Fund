```python
cluster_dic = {}
count = 0
for line in open("cluster.csv", "r"):
    for item in line.strip().split(","):
        cluster_dic[item] = count
    count += 1
```

```python
import time
import datetime
stage = {}
year_time = time.strftime('%Y', time.localtime(time.time()))
month_day = time.strftime('%m-%d', time.localtime(time.time()))
today = '{}-{}'.format(str(int(year_time)), month_day)
for i in range(1, 6, 2):
    end_time = '{}-{}'.format(str(int(year_time) - i), month_day)
    stage[i] = [end_time, today]
print(stage)
stage_2 = {}
for i in range(30,100,60):
    stage_2[i] = [(datetime.date.today() - datetime.timedelta(i)).strftime('%Y-%m-%d'), datetime.datetime.today().strftime('%Y-%m-%d')]
print(stage_2)
```

```python
info = {}
stage_improve = {}
for key in cluster_dic.keys():
    print(len(stage_improve))
    fund_em_open_fund_info_df = ak.fund_em_open_fund_info(fund=key, indicator="累计净值走势")
    stage_improve[key] = {}
    for sta, val in stage.items():
        start = fund_em_open_fund_info_df[fund_em_open_fund_info_df["净值日期"].astype(str) > val[0]]["累计净值"]
        end = fund_em_open_fund_info_df[fund_em_open_fund_info_df["净值日期"].astype(str) < val[1]]["累计净值"]
        start_val = 1.0
        end_val = 1.0
        if len(start) != 0:
            start_val = list(start)[0]
        if len(end) != 0:
            end_val = list(end)[-1]
        stage_improve[key][sta] = end_val / start_val
    for sta, val in stage_2.items():
        start = fund_em_open_fund_info_df[fund_em_open_fund_info_df["净值日期"].astype(str) > val[0]]["累计净值"]
        end = fund_em_open_fund_info_df[fund_em_open_fund_info_df["净值日期"].astype(str) < val[1]]["累计净值"]
        start_val = 1.0
        end_val = 1.0
        if len(start) != 0:
            start_val = list(start)[0]
        if len(end) != 0:
            end_val = list(end)[-1]
        stage_improve[key][sta] = end_val / start_val
```

```python
print(stage_improve)
```

```python
from collections import defaultdict
clusters = {}
count = 0
for line in open("cluster.csv", "r"):
    clusters[count] = line.strip().split(",")
    count += 1
stage_filter = {}
for sta in stage.keys():
    stage_filter[sta] = []
    for key, value in clusters.items():
        tmp = {}
        for val in value:
            tmp[val] = stage_improve[val][sta]
        l = sorted(tmp.items(), key=lambda d:d[1], reverse=True)
        stage_filter[sta] += [item[0] for item in l[:int(len(l)/6)]]
for sta in stage_2.keys():
    stage_filter[sta] = []
    for key, value in clusters.items():
        tmp = {}
        for val in value:
            tmp[val] = stage_improve[val][sta]
        l = sorted(tmp.items(), key=lambda d:d[1], reverse=True)
        stage_filter[sta] += [item[0] for item in l[:int(len(l)/8)]]
rule1 = set({})
for sta in stage.keys():
    if len(rule1) == 0:
        rule1 = set(stage_filter[sta])
    else:
        rule1 &= set(stage_filter[sta])
for sta in stage_2.keys():
    if len(rule1) == 0:
        rule1 = set(stage_filter[sta])
    else:
        rule1 &= set(stage_filter[sta])
print(len(rule1))
```
